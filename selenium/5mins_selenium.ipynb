```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

# création du pilote pour Chrome
driver = webdriver.Chrome(executable_path="/Users/assansanogo/Downloads/chromedriver")

# ouverture de la page associée à l’url via le pilote
driver.get("http://www.nike.com")
```

```python
# send key words
search_box = driver.find_element_by_css_selector("button.pre-search-btn.ripple")
search_box.send_keys("sneakers air force one customized", Keys.ENTER)
```

```python
# we identify the XPATH corresponding of the element of our choice
from selenium.webdriver.common.by import By
result = driver.find_element(By.XPATH, "//main/section/div/div[1]/div/figure/a[1]")
```

```python
# we identify all elements that follow our XPATH
elements = driver.find_elements_by_xpath("//div[@class='wall-image-loader css-1la3v4n ']//div//img")
links_list = [el.get_attribute("src") for el in elements]
print(links_list)
```
